<h4> Pochodna z definicji

```python
def pochodna(f, h = 0.0001):
    return lambda arg: (f(arg+h)-f(arg))/h

df = pochodna(lambda x: x**2)      
print(df(4))
```

```python
x = np.linspace(-5,5,100)
y = x**2*np.sin(x)                                  #funkcja
pochodna_analityczna = 2*x*np.sin(x)+x**2*np.cos(x) #policzona analitycznie
df = pochodna(lambda x: x**2*np.sin(x), 0.001) #pochodna obliczona z definicji
pochodna_przyblizona = np.array([df(i) for i in x]) 

plt.plot(x,y, c="blue")
plt.plot(x,pochodna_analityczna, c="green")
plt.plot(x,pochodna_przyblizona, c="red")

plt.show()
```

```python
from math import factorial

def sinus_taylor(x):
    y = 0
    n = 1
    for i in range(1,20,2):
        y += n*x**(i)/factorial(i)
        n = n*(-1)
    return y
```

```python
x = np.linspace(-4,4,100)
y = np.array([sinus_taylor(elem) for elem in x])
plt.plot(x,y , c ="blue")
plt.show()
```

```python
x = np.linspace(-9.5,9.5,100)
y = np.array([sinus_taylor(elem) for elem in x])
plt.plot(x,y , c ="blue")
plt.show()
```

<h4> Gdzie rozkład normalny pojawia się naturalnie w opisie zjawisk?

```python
#wykresy gęstosci dla rokzladu normlanego z roznymi parametrami

from math import pi, e

def f(x, mu = 0, sigma = 1):
    return (1/(sigma*(2*pi)**0.5))*e**(-((x-mu)**2)/(2*sigma**2))

x = [-6+0.1*i for i in range(200)]

y1 = [f(elem) for elem in x]  #N(0,1)
y2 = [f(elem, sigma = 2) for elem in x]   #N(0,2)
y3 = [f(elem, mu=3, sigma = 2) for elem in x]  #N(3,2)

plt.plot(x,y1, c = "red")
plt.plot(x,y2, c = "blue")
plt.plot(x,y3, c = "orange")
plt.show()
```

<h4> Centralne twierdzenie graniczne

```python
oczka = [1,2,3,4,5,6] #srednia=3.5

plt.hist(random.choices(oczka, k = 10000), bins='auto')
plt.show()
```

CTW: $\frac{\sum X_i - n\mu}{\sigma \sqrt{n}} \rightarrow N(0,1)$.

```python
def stand(y):
    n = len(y)
    return (sum(y)-n*3.5)/((2.917*n)**0.5)


y = []
for i in range(10000):
    y.append(stand(random.choices(oczka, k = 1000))) #ustandaryzowane sumy wynikow oczek przypominają dane z rozkladu N(0,1) - patrz nizej
    
plt.hist(y, bins = 100)
plt.show()
```

<h4> Twierdzenie Poissona

$\binom{n}{k}p^k(1-p)^{n-k} \rightarrow e^{-\lambda}\cdot \frac{\lambda^k}{k!}$, gdzie $\lambda \approx p\cdot n$.

```python
from math import comb, e, factorial

#rozklad dwupunktowy
def Binom(p,n,k):
    return comb(n,k)*(p**k)*((1-p)**(n-k))

#rozklad Poissona
def Poisson(lam, k):
    return e**(-lam) * lam**k / factorial(k)

#Jeżeli p - małe, i p*n stała liczba (gdy liczymy granice), to przyjmując lambda = p*n rozklady dwupunktowe i Poissona sa zblizone
Binom(1/100, 100, 3), Poisson(1/100*100, 3)
```
