```python
from CGRepresentation import CGR
from DFTree import dtree
from SSIMtree import stree
from Hurstree import htree
from LZtree import lztree
import random


nt = ["A"]*2+["T"]*2+["G"]*3+["C"]*3

def mutuj(seq, pt = 0.005, pd=0.001, pi = 0.001):
    """
    seq = sekwencja
    pt = prp tranzycji/transwersji
    pd = prp delecji
    pi = prp insercji
    """
    s = ""
    for i in range(len(seq)):
        r1 = random.uniform(0,1) #do tranzycji/transerwersji
        r2 = random.uniform(0,1) #do delecji
        r3 = random.uniform(0,1) #do insercji
        if r1 < pt:
            nuk = ["A"]*2+["T"]*2+["G"]*3+["C"]*3
            nuk = [elem for elem in nuk if elem != seq[i]]
            m = random.choices(nuk, k=1) #nowy nukleotyd
            s = s+m[0]
        if r2 < pd:
            s = s[:-1] #delecja ostatniego nt
        if r3 < pi:            
            nuk = ["A"]*2+["T"]*2+["G"]*3+["C"]*3
            m = random.choices(nuk, k=1) #nowy nukleotyd
            s = s + m[0]
        else:
            s = s + seq[i]
    return s
        
    
    
def przodki(seq, pt = 0.1, pd=0.05, pi = 0.05):
    """
    funkcja generuje 2 przodkow
    """
    m1 = mutuj(seq, pt, pd, pi)
    m2 = mutuj(seq, pt, pd, pi)
    return m1, m2

f = open("newicks.txt","w")

for n in [1000,10000,1000000]:
    for elem in [[0.005,0.001],[0.01,0.005],[0.1,0.01]]:
        pt, pd = elem
        pi = pd
        seq = random.choices(nt, k = n)
        seq = "".join(seq)
        
        sekwencje = {"seq":seq}
        for i in range(4):
            ss = {}
            for nazwy, elem in sekwencje.items():
                s1, s2 = przodki(elem, pt = pt, pd=pd, pi = pi)
                ss[nazwy+"1"] = s1
                ss[nazwy+"2"] = s2
            sekwencje = ss
            
        names = []
        sequences = []
        for a,b in sekwencje.items():
            names.append(a)
            sequences.append(b)
            
        print("Zaczynamy")
        
        t = lztree(sequences, names, seq_type = "DNA", tree_method = "ward")
        neww = t.newick()
        f.write(f"lz ward;{n};{pt};{pi};{neww}\n")
        
        t = lztree(sequences, names, seq_type = "DNA", tree_method = "average")
        neww = t.newick()
        f.write(f"lz average;{n};{pt};{pi};{neww}\n")
        
        t = dtree(sequences, names, seq_type = "DNA", tree_method = "ward")
        neww = t.newick()
        f.write(f"dft ward;{n};{pt};{pi};{neww}\n")
        
        t = dtree(sequences, names, seq_type = "DNA", tree_method = "average")
        neww = t.newick()
        f.write(f"dft average;{n};{pt};{pi};{neww}\n")
        
        t = htree(sequences, names, seq_type = "DNA", tree_method = "ward")
        neww = t.newick()
        f.write(f"Hurst ward;{n};{pt};{pi};{neww}\n")
        
        t = htree(sequences, names, seq_type = "DNA", tree_method = "average")
        neww = t.newick()
        f.write(f"Hurst average;{n};{pt};{pi};{neww}\n")
        
        t = stree(sequences, names, seq_type = "DNA", tree_method = "ward")
        neww = t.newick()
        f.write(f"SSIM ward;{n};{pt};{pi};{neww}\n")
        
        t = stree(sequences, names, seq_type = "DNA", tree_method = "average")
        neww = t.newick()
        f.write(f"SSIM average;{n};{pt};{pi};{neww}\n")
        
        print("To dziala!!")
f.close()
```
